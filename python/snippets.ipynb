```python
# pip install xlrd
import xlrd

# 打开Excel文件读取数据
data = xlrd.open_workbook_xls("./data/test.xls")     # 文件名以及路径，如果路径或者文件名有中文给前面加一个 r

# 获取book（excel文件）中一个工作表
# names = data.sheet_names()                  # 返回book中所有工作表的名字
table = data.sheets()[0]                    # 通过索引顺序获取
# table = data.sheet_by_index(sheet_indx)     # 通过索引顺序获取
# table = data.sheet_by_name(sheet_name)      # 通过名称获取
# data.sheet_loaded(sheet_name or indx)       # 检查某个sheet是否导入完毕
# print(table)

print("-" * 50)
# 获取该sheet中的行数，注，这里table.nrows后面不带().
nrows = table.nrows
print(nrows)

# 返回由该行中所有的单元格对象组成的列表,这与tabel.raw()方法并没有区别。
rowx_objlist = table.row(0)
print(rowx_objlist)

# 返回由该行中所有的单元格对象组成的列表
rowx_objlist = table.row_slice(1)
print(rowx_objlist)

# 返回由该行中所有单元格的数据类型组成的列表；　　　　
# 返回值为逻辑值列表，若类型为empy则为0，否则为1
rowx_typelist = table.row_types(1, start_colx=0, end_colx=None)
print(rowx_typelist)

# 返回由该行中所有单元格的数据组成的列表
rowx_vallist = table.row_values(2, start_colx=0, end_colx=None)
print(rowx_vallist)

# 返回该行的有效单元格长度，即这一行有多少个数据
rowx_len = table.row_len(2)
print(rowx_len)

print("-" * 50)


print("-" * 50)
# 获取列表的有效列数
ncols = table.ncols
print(ncols)

# 返回由该列中所有的单元格对象组成的列表
colx_objlist = table.col(0, start_rowx=0, end_rowx=None)
print(colx_objlist)

# 返回由该列中所有的单元格对象组成的列表
colx_objlist = table.col_slice(1, start_rowx=0, end_rowx=None)
print(colx_objlist)

# 返回由该列中所有单元格的数据类型组成的列表
colx_typelist = table.col_types(2, start_rowx=0, end_rowx=None)
print(colx_typelist)

# 返回由该列中所有单元格的数据组成的列表
colx_vallist = table.col_values(1, start_rowx=0, end_rowx=None)
print(colx_vallist)

print("-" * 50)


print("-" * 50)
# 返回单元格对象
cellx_obj = table.cell(2, 0)  # A2
print(cellx_obj)

# 返回对应位置单元格中的数据类型
cellx_type = table.cell_type(2, 1)
print(cellx_type)

# 返回对应位置单元格中的数据
cellx_val = table.cell_value(2, 2)
print(cellx_val)

print("-" * 50)
```

```python
# pip install xlwt
import xlwt

workbook = xlwt.Workbook(encoding='utf-8')
# 创建新的sheet表
worksheet = workbook.add_sheet("New Sheet")

# 设置列宽
worksheet.col(0).width = 256*20   # xlwt中列宽的值表示方法：默认字体0的1/256为衡量单位
# 设置行高
row_style = xlwt.easyxf('font:height 360;')  # 18pt,类型小初的字号
row = worksheet.row(0)
row.set_style(row_style)


# 初始化样式
style = xlwt.XFStyle()

# 设置字体样式
font = xlwt.Font()
font.name = 'Times New Roman'  # 字体
font.bold = True               # 加粗
font.underline = True          # 下划线
font.italic = True             # 斜体
# 设置样式
style.font = font

# 设置边框样式
# May be:   NO_LINE, THIN, MEDIUM, DASHED, DOTTED, THICK, DOUBLE, HAIR,
#           MEDIUM_DASHED, THIN_DASH_DOTTED, MEDIUM_DASH_DOTTED, THIN_DASH_DOT_DOTTED,
#           MEDIUM_DASH_DOT_DOTTED, SLANTED_MEDIUM_DASH_DOTTED, or 0x00 through 0x0D.
# DASHED虚线
# NO_LINE没有
# THIN实线
borders = xlwt.Borders()  # Create Borders
borders.left = xlwt.Borders.DASHED
borders.right = xlwt.Borders.DASHED
borders.top = xlwt.Borders.DASHED
borders.bottom = xlwt.Borders.DASHED
borders.left_colour = 0x40
borders.right_colour = 0x40
borders.top_colour = 0x40
borders.bottom_colour = 0x40
# Add Borders to Style
style.borders = borders

# 设置背景样式
pattern = xlwt.Pattern()
# May be: NO_PATTERN, SOLID_PATTERN, or 0x00 through 0x12
pattern.pattern = xlwt.Pattern.SOLID_PATTERN

# May be: 8 through 63. 0 = Black, 1 = White, 2 = Red, 3 = Green, 4 = Blue, 5 = Yellow,
# 6 = Magenta, 7 = Cyan, 16 = Maroon, 17 = Dark Green, 18 = Dark Blue, 19 = Dark Yellow ,
# almost brown), 20 = Dark Magenta, 21 = Teal, 22 = Light Gray, 23 = Dark Gray, the list goes on...
pattern.pattern_fore_colour = 5
style.pattern = pattern

# 设置对齐样式
al = xlwt.Alignment()
# VERT_TOP = 0x00       上端对齐
# VERT_CENTER = 0x01    居中对齐（垂直方向上）
# VERT_BOTTOM = 0x02    低端对齐
# HORZ_LEFT = 0x01      左端对齐
# HORZ_CENTER = 0x02    居中对齐（水平方向上）
# HORZ_RIGHT = 0x03     右端对齐
al.horz = 0x02  # 设置水平居中
al.vert = 0x01  # 设置垂直居中
style.alignment = al


# 往表格写入内容
worksheet.write(0, 0, "内容1")
worksheet.write(2, 1, "内容2", style)  # 使用样式


# 合并第1行到第2行的第0列到第3列
# worksheet.write_merge(1, 2, 0, 3, 'Merge Test')

# 保存
workbook.save("./data/new_test.xls")
```

## read and write xml file

```python
import xml.dom.minidom

xml_global_dict = {}
# 使用minidom解析器打开 XML 文档
dom = xml.dom.minidom.parse("./data/configuration.xml")

root_node = dom.documentElement  # 获取根节点
audioswitch_node = root_node.getElementsByTagName('audioswitch')[0]
devicesconf_node = root_node.getElementsByTagName('devicesconf')[0]
alsaportsconf_node = root_node.getElementsByTagName('alsaportsconf')[0]
moduleslist_node = root_node.getElementsByTagName('moduleslist')[0]
pathconf_node = root_node.getElementsByTagName('pathconf')[0]
# xi_include_node = root_node.getElementsByTagName('xi:include')[0]

switch_node_list = audioswitch_node.getElementsByTagName('switch')
device_node_list = devicesconf_node.getElementsByTagName('device')
port_node_list = alsaportsconf_node.getElementsByTagName('port')
module_node_list = moduleslist_node.getElementsByTagName('module')
path_node_list = pathconf_node.getElementsByTagName('path')


xml_global_dict["audioRoutingConfiguration"] = {}
audioswitch_list = []
devicesconf_list = []
alsaportsconf_list = []
moduleslist_list = []
pathconf_list = []

for item in switch_node_list:
    audioswitch_list.append({"switch": {k: v for k, v in item.attributes.items()}})
xml_global_dict["audioRoutingConfiguration"].update(audioswitch=audioswitch_list)
# print(f"audioswitch_list: {audioswitch_list}")
# print(f"xml_global_dict: {xml_global_dict}")

for item in device_node_list:
    devicesconf_list.append({"device": {k: v for k, v in item.attributes.items()}})
xml_global_dict["audioRoutingConfiguration"].update(devicesconf=devicesconf_list)
# print(f"devicesconf_list: {devicesconf_list}")
# print(f"xml_global_dict: {xml_global_dict}")

for item in port_node_list:
    alsaportsconf_list.append({"port": {k: v for k, v in item.attributes.items()}})
xml_global_dict["audioRoutingConfiguration"].update(devicesconf=alsaportsconf_list)
# print(f"alsaportsconf_list: {alsaportsconf_list}")
# print(f"xml_global_dict: {xml_global_dict}")

for item in module_node_list:
    moduleslist_list.append({"module": {k: v for k, v in item.attributes.items()}})
xml_global_dict["audioRoutingConfiguration"].update(moduleslist=moduleslist_list)
# print(f"moduleslist_list: {moduleslist_list}")
# print(f"xml_global_dict: {xml_global_dict}")
```

```python
import xml.etree.ElementTree as ET

global_dict = {"xml_file": "./data/configuration.xml",
               "audioswitch_node": "audioswitch",
               "devicesconf_node": "devicesconf",
               "alsaportsconf_node": "alsaportsconf",
               "moduleslist_node": "moduleslist",
               "pathconf_node": "pathconf",
               "switch_node": "audioswitch/switch",
               "device_node": "devicesconf/device",
               "port_node": "alsaportsconf/port",
               "module_node": "moduleslist/module",
               "path_node": "pathconf/path",
               "connections_node": "connections",
               "connection_node": "connections/connection",
               "includes_node": "includes",
               "include_node": "includes/include"
               }
xml_global_dict = {}

tree = ET.parse(global_dict["xml_file"])
root = tree.getroot()
xml_global_dict = {root.tag: {}}
# root_tag_name = root.tag
audioswitch_node = tree.find(global_dict["audioswitch_node"])
devicesconf_node = tree.find(global_dict["devicesconf_node"])
alsaportsconf_node = tree.find(global_dict["alsaportsconf_node"])
moduleslist_node = tree.find(global_dict["moduleslist_node"])
pathconf_node = tree.find(global_dict["pathconf_node"])
switch_node_list = tree.findall(global_dict["switch_node"])
device_node_list = tree.findall(global_dict["device_node"])
port_node_list = tree.findall(global_dict["port_node"])
module_node_list = tree.findall(global_dict["module_node"])
path_node_list = tree.findall(global_dict["path_node"])

xml_global_dict[root.tag][audioswitch_node.tag] = []
for item in switch_node_list:
    xml_global_dict[root.tag][audioswitch_node.tag].append({item.tag: item.attrib})
# print(xml_global_dict)

xml_global_dict[root.tag][devicesconf_node.tag] = []
for item in device_node_list:
    xml_global_dict[root.tag][devicesconf_node.tag].append({item.tag: item.attrib})
# print(xml_global_dict)

xml_global_dict[root.tag][alsaportsconf_node.tag] = []
for item in port_node_list:
    xml_global_dict[root.tag][alsaportsconf_node.tag].append({item.tag: item.attrib})
# print(xml_global_dict)

xml_global_dict[root.tag][moduleslist_node.tag] = []
for item in module_node_list:
    xml_global_dict[root.tag][moduleslist_node.tag].append({item.tag: item.attrib})
# print(xml_global_dict)

xml_global_dict[root.tag][pathconf_node.tag] = []
for i, item in enumerate(path_node_list):
    if item.find(global_dict["connections_node"]):
        xml_global_dict[root.tag][pathconf_node.tag].append(
            {item.tag: {"attrib": item.attrib, global_dict["connections_node"]: []}})
        for s_item in item.findall(global_dict["connection_node"]):
            xml_global_dict[root.tag][pathconf_node.tag][i][item.tag][global_dict["connections_node"]].append({
                                                                                                              s_item.tag: s_item.attrib})
    elif item.find(global_dict["includes_node"]):
        xml_global_dict[root.tag][pathconf_node.tag].append(
            {item.tag: {"attrib": item.attrib, global_dict["includes_node"]: []}})
        for s_item in item.findall(global_dict["include_node"]):
            xml_global_dict[root.tag][pathconf_node.tag][i][item.tag][global_dict["includes_node"]].append({
                                                                                                           s_item.tag: s_item.attrib})
print(xml_global_dict)
```
